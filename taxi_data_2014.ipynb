```python
taxi_data[" pickup_datetime"] = pd.to_datetime(taxi_data[" pickup_datetime"], infer_datetime_format=True)
```

```python
# we should sort the taxi data by date, as we likely need this for our time estimation algorithm
taxi_data = taxi_data.sort_values(" pickup_datetime")
```

```python
taxi_data
```

```python
# lets do 10% of traffic in January 10, from 8AM-11PM, for example
date1 = '2014-01-10 08:00:00'
date2 = '2014-01-10 23:00:00'
small_data = taxi_data[(taxi_data[' pickup_datetime'] > date1) & (taxi_data[' pickup_datetime'] <= date2)]
small_data = small_data[:len(small_data)]
```

```python
class TaxiEntry:
    def __init__(self, t_pickup, t_dropoff, long_pickup, lat_pickup, long_dropoff, lat_dropoff):
        self.t_pickup = t_pickup
        self.t_dropoff = t_dropoff
        self.long_pickup, self.lat_pickup = long_pickup, lat_pickup
        self.long_dropoff, self.lat_dropoff = long_dropoff, lat_dropoff
```

```python
# let's split by hour
group_by_hour = {}

for index, val in small_data.iterrows():
    pickup_time = val[" pickup_datetime"]
    dropoff_time = val[" dropoff_datetime"]
    pickup_long, pickup_lat = val[" pickup_longitude"], val[" pickup_latitude"]
    dropoff_long, dropoff_lat = val[" dropoff_longitude"], val[" dropoff_latitude"]
    
    new_taxi_entry = TaxiEntry(pickup_time, dropoff_time, pickup_long, pickup_lat, dropoff_long, dropoff_lat)
    hour = pickup_time.hour
    group_by_hour[hour] = group_by_hour.get(hour, []) + [new_taxi_entry]
```

```python
rides_by_hour = {k:len(v) for k,v in group_by_hour.items()}
rides_by_hour
```
